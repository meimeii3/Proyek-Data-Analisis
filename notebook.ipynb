### Cleaning Data

### Menangani Missing Value

```python
# Temp, Press,Dewp,Rain,wspm memiliki sedikit missing values sehingga akan digunakan imputation
df['TEMP'].fillna(df['TEMP'].mean(), inplace=True)
df['PRES'].fillna(df['PRES'].mean(), inplace=True)
df['DEWP'].fillna(df['DEWP'].mean(), inplace=True)
df['RAIN'].fillna(df['RAIN'].mean(), inplace=True)
df['WSPM'].fillna(df['WSPM'].mean(), inplace=True)
```

```python
# Mengisi dengan interpolasi untuk kolom yang missing value lebih besar
df['PM2.5'].interpolate(method='linear', inplace=True)
df['PM10'].interpolate(method='linear', inplace=True)
df['SO2'].interpolate(method='linear', inplace=True)
df['NO2'].interpolate(method='linear', inplace=True)
df['CO'].interpolate(method='linear', inplace=True)
df['O3'].interpolate(method='linear', inplace=True)
```

```python
# Mengisi kolom wd dengan mode (karena data kategorikal)
df['wd'].fillna(df['wd'].mode()[0], inplace=True)
```

```python
df.isna().sum()
```

```python
# Mengubah kolom year, month dan day menjadi kolom datetime
df['datetime'] = pd.to_datetime(df[['year', 'month', 'day','hour']])
df = df.drop(columns=['year', 'month', 'day', 'hour'])
```

## Analisis Lanjutan

```python
# Fungsi untuk mengelompokkan berdasarkan PM2.5 dan PM10
def pollution_cluster(row):
    if row['PM2.5'] <= 50 and row['PM10'] <= 50:
        return 'Low Pollution'
    elif 50 < row['PM2.5'] <= 150 or 50 < row['PM10'] <= 150:
        return 'Moderate Pollution'
    else:
        return 'High Pollution'

# Fungsi untuk mengelompokkan berdasarkan CO
def co_cluster(row):
    if row['CO'] <= 1000:
        return 'Low CO'
    elif 1000 < row['CO'] <= 5000:
        return 'Moderate CO'
    else:
        return 'High CO'

# Fungsi untuk mengelompokkan berdasarkan suhu (TEMP)
def temp_cluster(row):
    if row['TEMP'] < 0:
        return 'Cold'
    elif 0 <= row['TEMP'] <= 20:
        return 'Mild'
    else:
        return 'Hot'

# Terapkan fungsi clustering pada dataframe
df['Pollution Cluster'] = df.apply(pollution_cluster, axis=1)
df['CO Cluster'] = df.apply(co_cluster, axis=1)
df['Temperature Cluster'] = df.apply(temp_cluster, axis=1)

# Lihat beberapa baris dari hasil clustering
df[['PM2.5', 'PM10', 'CO', 'TEMP', 'Pollution Cluster', 'CO Cluster', 'Temperature Cluster']]

```

```python
# Menghitung distribusi tiap cluster
pollution_distribution = df['Pollution Cluster'].value_counts()
co_distribution = df['CO Cluster'].value_counts()
temp_distribution = df['Temperature Cluster'].value_counts()

print("Pollution Cluster Distribution:\n", pollution_distribution)
print("\nCO Cluster Distribution:\n", co_distribution)
print("\nTemperature Cluster Distribution:\n", temp_distribution)

```

**insight**
- Moderate Pollution : 216,042 adalah kategori terbesar, menunjukkan bahwa lebih dari setengah data berada dalam rentang polusi yang moderat yang menunjukkan bahwa pada banyak waktu pengukuran, kualitas udara cenderung menengah.
- Low CO: 245,501 sebagian besar data menunjukkan tingkat karbon monoksida yang rendah
- Mild: 206,612 hampir setengah dari dataset menunjukkan suhu sedang
